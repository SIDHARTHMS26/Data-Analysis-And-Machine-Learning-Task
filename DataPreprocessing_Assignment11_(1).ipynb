```python
data=pd.read_csv("/content/drive/MyDrive/Task 2/Assignment11_to_16/Data.csv")
print(data.shape)
data.head()
```

**Step 3: Handling the missing data**

```python
data.dropna(inplace=True)
data
```

**Step 4: Encoding categorical data**

```python
from sklearn import preprocessing
le=preprocessing.LabelEncoder()
data['Country_enc']=le.fit_transform(data['Country'])
data['Purchased_enc']=le.fit_transform(data['Purchased'])
data.head()
```

**Step 5: Creating a dummy variable**

```python
pd.get_dummies(data)
```

**Step 6: Splitting the datasets into training sets and Test sets**

```python
x=data[['Country_enc','Age','Salary']]
y=data['Purchased_enc']
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.4,random_state=0)
```

**Step 7: Feature Scaling**

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
pipeline=Pipeline([
                   ('std_scalar',StandardScaler())
])
xtrain=pipeline.fit_transform(xtrain)
xtest=pipeline.fit_transform(xtest)
```
